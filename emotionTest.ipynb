```python
from keras.models import load_model
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random

model = load_model('my_model.h5')

data = pd.read_csv('icml_face_data.csv')

random_indices = random.sample(range(data.shape[0]), 10)

for i, idx in enumerate(random_indices):

    pixels = data[' pixels'].iloc[idx]

    img_tensor = np.fromstring(pixels, sep=' ').reshape(48, 48, 1)

    img_tensor = np.expand_dims(img_tensor, axis=0)

    img_tensor_normalized = img_tensor.astype('float32') / 255.

    prediction = model.predict(img_tensor_normalized)

    emotion_labels = ["Angry", "Disgust", "Fear", "Happy", "Sad", "Surprise", "Neutral"]

    fig, axs = plt.subplots(1, 2, figsize=(12, 6))

    axs[0].imshow(img_tensor[0, :, :, 0], cmap='gray', interpolation='bicubic')
    axs[0].set_title('Input Image')

    sns.barplot(x=emotion_labels, y=prediction[0], palette="viridis", ax=axs[1])
    axs[1].set_title('Emotion Prediction')
    axs[1].set_xlabel('Emotions')
    axs[1].set_ylabel('Probability')

    plt.show()
```

```python
from tensorflow.keras.preprocessing import image
from tensorflow.keras.models import load_model
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Modeli yükle
model = load_model('my_model.h5')

# Görüntüyü yükle
img = image.load_img('ekrem.jpg', color_mode='grayscale', target_size=(48, 48))

# Görüntüyü bir numpy dizisine dönüştür
img_tensor = image.img_to_array(img)
img_tensor = np.expand_dims(img_tensor, axis=0)

# Görüntüyü normalize et
img_tensor /= 255.

# Tahmin yap
prediction = model.predict(img_tensor)

# Tahmin sonucunu görselleştir
emotion_labels = ["Angry", "Disgust", "Fear", "Happy", "Sad", "Surprise", "Neutral"]

# Çizim için bir figure ve axes oluştur
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# Görüntüyü çiz
axs[0].imshow(img_tensor[0, :, :, 0], cmap='gray', interpolation='bicubic')
axs[0].set_title('Ufes Sogut')

# Çubuk grafiği çiz
sns.barplot(x=emotion_labels, y=prediction[0], palette="viridis", ax=axs[1])
axs[1].set_title('Emotion Prediction')
axs[1].set_xlabel('Emotions')
axs[1].set_ylabel('Probability')

plt.show()
```

```python
from tensorflow.keras.preprocessing import image
from tensorflow.keras.models import load_model
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Modeli yükle
model = load_model('my_model.h5')

# Görüntüyü yükle
img = image.load_img('ekrem2.jpg', color_mode='grayscale')

# Görüntüyü daha büyük bir boyuta getir
img = img.resize((96, 96))

# Görüntüyü modelin beklediği boyuta getir
img = img.resize((48, 48))

# Görüntüyü bir numpy dizisine dönüştür
img_tensor = image.img_to_array(img)
img_tensor = np.expand_dims(img_tensor, axis=0)

# Görüntüyü normalize et
img_tensor /= 255.

# Tahmin yap
prediction = model.predict(img_tensor)

# Tahmin sonucunu görselleştir
emotion_labels = ["Angry", "Disgust", "Fear", "Happy", "Sad", "Surprise", "Neutral"]

# Çizim için bir figure ve axes oluştur
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# Görüntüyü çiz
axs[0].imshow(img_tensor[0, :, :, 0], cmap='gray', interpolation='bicubic')
axs[0].set_title('Input Image')

# Çubuk grafiği çiz
sns.barplot(x=emotion_labels, y=prediction[0], palette="viridis", ax=axs[1])
axs[1].set_title('Emotion Prediction')
axs[1].set_xlabel('Emotions')
axs[1].set_ylabel('Probability')

plt.show()
```
